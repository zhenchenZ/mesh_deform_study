```python
from mesh_tools import load_mesh, analyze_mesh, quick_plots, trellis_like_deform
import numpy as np
import open3d as o3d

# 1) Load CAD mesh + labels (face_labels: np.ndarray of shape [n_faces])
mesh = load_mesh("cad_example.obj")
face_labels = np.load("cad_example_face_labels.npy")  # 0=plane,1=cylinder,2=other

# 2) Forensics on original CAD
summary_cad = analyze_mesh(mesh, face_labels=face_labels)
print(summary_cad)

# 3) Generate trellis-like version
mesh_trellis = trellis_like_deform(mesh, face_labels=face_labels)

# 4) Forensics on trellis version
summary_trellis = analyze_mesh(mesh_trellis, face_labels=face_labels)
print(summary_trellis)

# 5) Quick plots in notebook
quick_plots(mesh)
quick_plots(mesh_trellis)

# 6) Visualize in Open3D if you want
o3d.visualization.draw_geometries([mesh_trellis])
```
